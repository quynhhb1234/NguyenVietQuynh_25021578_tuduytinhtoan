```python
#w6a1
num=list(map(int, input().split()))
seen=set()
result=[]
for x in num:
    if x not in seen:
        result.append(x)
        seen.add(x)
print(result)
```

```python
#w6a2
lis=list(map(int, input().split()))
result=[]
tong=0
for x in lis:
    tong+=x
    result.append(tong)
print(result)
```

```python
#w6a4
lis= list(map(int, input().split()))
dic={"positives": 0, "negatives": 0, "zeros": 0}
for x in lis:
    if x>0: dic["positives"] +=1
    elif x<0: dic["negatives"] +=1
    else: dic["zeros"] +=1
print(dic)
    
```
