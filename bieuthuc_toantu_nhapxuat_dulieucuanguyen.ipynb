```python
#bai 9
x = float(input("Nhập giá đồng hồ (USD): "))
shipping_fee = 10
subtotal = x + shipping_fee
import_tax = 0.3 * subtotal
vat_tax = 0.1 * subtotal
total = subtotal + import_tax + vat_tax
print(f"Tổng số tiền phải trả: {total:.2f} USD")
```
