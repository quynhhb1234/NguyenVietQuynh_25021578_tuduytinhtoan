```python
a = [4, 3, 2, 6, 7, 1, 8, 9, 10, 0, 5, 8, 9]
n = len(a)

dp = [1]*n

for i in range(n):
    for j in range(i):
        if a[j] < a[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
print(max(dp))
```
