Giải thích từng dòng:
	1.	a = 7
Gán (lưu) giá trị số nguyên 7 vào biến tên a. Nghĩa là từ đây trong chương trình, khi bạn gọi a thì nó đại diện cho số 7.
	2.	b = 5
Gán giá trị số nguyên 5 vào biến b. Tương tự, b giờ đại diện cho số 5.
	3.	c = a - b
Thực hiện phép trừ: lấy giá trị của a trừ đi giá trị của b. Vì a là 7 và b là 5, nên a - b = 7 - 5. Tính từng chữ số: 7 trừ 5 bằng 2. Kết quả phép tính được lưu vào biến c, vậy c bây giờ chứa giá trị 2.
	4.	print(c)
Lệnh in (xuất) giá trị hiện có trong biến c ra màn hình/Terminal. Vì c là 2, nên chương trình sẽ in ra 2.

Tóm tắt: chương trình khởi tạo hai biến a = 7 và b = 5, tính hiệu của chúng (7 − 5) và in kết quả. Kết quả là 2 — mình đã chạy đoạn code thực tế và Terminal trả về 2 như thấy ở trên.

```python
a = 7
b = 5
c = a - b
print(c)
```

Giải thích từng dòng:
	1.	city = "Hà Nội"
Bạn tạo một biến tên city và gán cho nó một chuỗi ký tự. Chuỗi này là "Hà Nội". Bất cứ lúc nào bạn gọi city, Python sẽ hiểu đó là chữ Hà Nội.
	2.	year = 2025
Dòng này tạo biến year và gán cho nó giá trị số nguyên 2025. Không cần dấu ngoặc kép vì đây là số, không phải chuỗi.
	3.	print("Thành phố: ", city, " - Năm: ", year)
Lệnh print in ra màn hình tất cả những gì nó nhận được.
Python sẽ nối (ghép) các phần này theo đúng thứ tự:
• chuỗi "Thành phố: "
• giá trị của biến city (tức “Hà Nội”)
• chuỗi " - Năm: "
• giá trị của biến year (2025)
Khi ghép lại, Python cũng tự chèn khoảng trắng giữa các tham số trong print.


```python
city = "Hà Nội" 
year = 2025 
print("Thành phố: ", city, " - Năm: ", year)
```

Giải thích từng dòng:
	1.	n = 4
Bạn đặt biến n bằng 4. Nghĩa là bạn muốn tính tổng từ 1 đến 4.
	2.	t = 0
Tạo biến t để chứa tổng. Lúc đầu tổng bằng 0 để bắt đầu cộng dồn.
	3.	for i in range(1, n+1):
Vòng lặp chạy biến i từ 1 đến n (tức là 4).
range(1, n+1) nghĩa là chạy i = 1, 2, 3, 4.
	4.	t += i
Mỗi vòng lặp, cộng giá trị của i vào t.
Chạy từng vòng:
• Lần 1: t = 0 + 1 = 1
• Lần 2: t = 1 + 2 = 3
• Lần 3: t = 3 + 3 = 6
• Lần 4: t = 6 + 4 = 10
	5.	print(t)
In ra tổng cuối cùng.
