```python
#w5a1
def Max(a,b) -> int:
    if a>b:
        return a
    else:
        return b
a,b= map(int, input().split())
print(Max(a,b))
```

```python
#w5a2
def swap(a:int,b:int)-> int:
    a,b=b,a
    return a,b
a,b=map(int, input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#w5a3
def is_prime(n:int) -> bool :
    if n<2:
        return False 
    for i in range (2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
n=int(input())
print(is_prime(n))
        
```

```python
#w5a4
def hoanhoa(n:int) -> bool:
    if n<=1:
        return False
    tong = 0
    for i in range(1,n):
        if n % i == 0:
            tong += i
    return tong == n
n =int(input())
print(hoanhoa(n))
```

```python
#w5a5
def sodau(lst: list[int],k:int)-> int:
    for i in range(len(lst)):
        if lst [i] == k:
            return i+1
    return -1
lst = list (map(int, input().split()))
k=int(input())
print(sodau(lst,k))
```

```python
#w5a6
def giathua(n:int) -> int:
    tich=1
    for i in range (1, n+1):
        tich *= i
    return tich
n =int(input())
print(giathua(n))
```
