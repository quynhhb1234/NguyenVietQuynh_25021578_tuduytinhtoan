```python
#w5a1
def Max(a,b) -> int:
    if a>b:
        return a
    else:
        return b
a,b= map(int, input().split())
print(Max(a,b))
```

```python
#w5a2
def swap(a:int,b:int)-> int:
    a,b=b,a
    return a,b
a,b=map(int, input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#w5a3
def is_prime(n:int) -> bool :
    if n<2:
        return False 
    for i in range (2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
n=int(input())
print(is_prime(n))
        
```

```python
#w5a4
def hoanhoa(n:int) -> bool:
    if n<=1:
        return False
    tong = 0
    for i in range(1,n):
        if n % i == 0:
            tong += i
    return tong == n
n =int(input())
print(hoanhoa(n))
```

```python
#w5a5
def sodau(lst: list[int],k:int)-> int:
    for i in range(len(lst)):
        if lst [i] == k:
            return i+1
    return -1
lst = list (map(int, input().split()))
k=int(input())
print(sodau(lst,k))
```

```python
#w5a6
def giathua(n:int) -> int:
    tich=1
    for i in range (1, n+1):
        tich *= i
    return tich
n =int(input())
print(giathua(n))
```

```python
#w5a7
def maytinh(num1: float,num2:float) :
    tong = num1 + num2 
    hieu = num1 - num2
    tich = num1 * num2
    thuong = num1 / num2 if num2 != 0 else "lỗi"
    return tong,hieu,tich,thuong
num1, num2 = map(float, input().split())
tong,hieu,tich,thuong= maytinh(num1,num2)
print(f"tổng: {tong}")
print(f"hiệu: {hieu}")
print(f"tích: {tich}")
print(f"thương: {thuong}")
```

```python
#w5a8
def hamming(x:int , y:int):
    xor = x^y
    return bin(xor).count('1')
x,y=map(int, input().split())
print(hamming(x,y))
```

```python
#w5a9
def tong(n:int):
    s=0
    for i in range (1, n+1):
        while n>0:
            s += n % 10
            n //= 10
            return s
n=int(input())
print(tong(n))
```
