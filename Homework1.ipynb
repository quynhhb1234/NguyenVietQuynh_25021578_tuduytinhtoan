bài 2: a=3 b=10

```python
a, b = map(float, input().split())

r = min(a, b) / 2
area_rect = a * b
area_circle = 3.14 * r * r

result = area_rect - area_circle
print(f"{result:.2f}")
```

bài 5: c là a

```python
c = input()

# giả sử chắc chắn người dùng nhập chữ hoa
if c == 'A':
    print("Không có chữ cái đứng trước a")
else:
    lower = c.lower()              # đổi sang chữ thường
    prev_char = chr(ord(lower) - 1)  # lùi 1 kí tự
    print(prev_char)
```

bài 6: a=3 b=4 c=5

```python
a = float(input())
b = float(input())
c = float(input())

# điều kiện kiểm tra tam giác
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2          # nửa chu vi
    s = (p*(p-a)*(p-b)*(p-c)) ** 0.5   # công thức Heron
    print(f"{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
name = input("Ho ten chu ho: ")
old = int(input("Chi so thang truoc: "))
new = int(input("Chi so thang nay: "))

d = new - old   # số kWh tiêu thụ
tien = 0

if d <= 50:
    tien = d * 1984
elif d <= 100:
    tien = 50 * 1984 + (d - 50) * 2050
elif d <= 200:
    tien = 50 * 1984 + 50 * 2050 + (d - 100) * 2380
elif d <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (d - 200) * 2998
elif d <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (d - 300) * 3350
else:
    tien = (50 * 1984 + 50 * 2050 + 100 * 2380 + 
            100 * 2998 + 100 * 3350 + (d - 400) * 3460)

# VAT 8%
tien = int(tien * 1.08)
print("họ và tên chủ hộ:", name)
print("Tien dien phai tra:", tien, "dong")
```
